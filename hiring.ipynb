```python
df=pd.read_csv("hiring.csv")
df
```

```python
df.experience=df.experience.fillna("Zero")
df
```

```python
df.experience = df.experience.apply(w2n.word_to_num)
df
```

# Fillin the NAN values of Experience coloumn by number "0" by using word2num ibrary.
df.experience = df.experience.apply(w2n.word_to_num)<br>
df

```python
import math
median_test_score=math.floor(df['test_score(out of 10)'].mean())
median_test_score
```

```python
df['test_score(out of 10)']=df['test_score(out of 10)'].fillna(median_test_score)
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']],df[['salary($)']])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2,9,6]])
```

```python
reg.predict([[12,10,10]])
```

# Saving the ML model by using **Joblib Library** of Python in oreder to make it easy to be used by others just by sending the model_joblib file . We don't need to repeatedly tain the whole dataset for predicting the value of the salary

```python
joblib.dump(reg,'model_joblib')
```

```python
mj=joblib.load('model_joblib')
```

```python
mj.predict([[2,9,6]])
```

```python
mj.coef_
```

# Saving the ML model by using **Pickle Library** of Python in oreder to make it easy to be used by others just by sending the model_joblib file . We don't need to repeatedly tain the whole dataset for predicting the value of the salary

```python
import pickle
```

```python
with open('model_pickle','wb') as f:
  pickle.dump(reg,f)
```

```python
with open('model_pickle','rb') as f:
  mp=pickle.load(f)
```

```python
mp.predict([[2,4,6]])
```
